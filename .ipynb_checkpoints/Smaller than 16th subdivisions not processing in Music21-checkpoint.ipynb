```python
BeethovenEx.measures(0, 2).show('midi')
```

```python
newBeethovenObject = BeethovenEx.measures(0, 2)
newBeethovenMidi = midi.translate.streamToMidiFile(newBeethovenObject)
```

```python
newBeethovenMidi.open('rewrittenBeethoven.mid', 'wb')
newBeethovenMidi.write()
newBeethovenMidi.close()
```

```python
BeethovenTest = converter.parse('rewrittenBeethoven.mid')
```
